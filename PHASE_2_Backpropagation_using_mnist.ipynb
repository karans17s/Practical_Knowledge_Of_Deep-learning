```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# 1. Load the training and testing data
train_data = pd.read_csv('/content/mnist_train.csv')  # MNIST training dataset
test_data = pd.read_csv('/content/mnist_test.csv')    # MNIST testing dataset

# 2. Extract features (pixel values) and labels (digits)
x = train_data.iloc[:, 1:].values  # Features: all columns except the first
y = train_data.iloc[:, 0].values   # Labels: the first column

# 3. Normalize pixel values to [0, 1] for faster convergence
x = x / 255.0

# 4. Split training data into training and validation sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# 5. Standardize data to improve training stability
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)  # Fit on training data and transform
x_test = scaler.transform(x_test)        # Transform test data using the same scaler

# 6. Define and train a neural network (MLP Classifier)
mlp = MLPClassifier(
    hidden_layer_sizes=(128,),     # One hidden layer with 128 neurons
    activation='logistic',         # Logistic sigmoid activation function
    max_iter=100,                  # Train for 100 iterations
    verbose=True                   # Print training progress
)
mlp.fit(x_train, y_train)          # Train the model

# 7. Evaluate the model on the test set
y_pred = mlp.predict(x_test)       # Predict labels for test data
accuracy = accuracy_score(y_test, y_pred)  # Calculate accuracy
print(f'Accuracy: {accuracy * 100:.2f}%')

# 8. Plot the loss curve to visualize training progress
plt.plot(mlp.loss_curve_, color='red')   # Loss values over iterations
plt.title('Loss Curve during Training')  # Title of the plot
plt.xlabel('Iterations')                 # X-axis: training iterations
plt.ylabel('Loss')                       # Y-axis: loss
plt.show()
```

```python
import numpy as np
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# 1. Load MNIST dataset directly from OpenML
mnist = fetch_openml('mnist_784', version=1, as_frame=False)  # Load MNIST data
x, y = mnist.data, mnist.target.astype(int)  # Features and Labels

# 2. Normalize pixel values to [0, 1]
x = x / 255.0

# 3. Split data into training and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# 4. Standardize data
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

# 5. Define and train a neural network
mlp = MLPClassifier(
    hidden_layer_sizes=(128,),     # One hidden layer with 128 neurons
    activation='logistic',         # Logistic sigmoid activation function
    max_iter=100,                  # Train for 100 iterations
    verbose=True                   # Print training progress
)
mlp.fit(x_train, y_train)          # Train the model

# 6. Evaluate the model
y_pred = mlp.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')

# 7. Plot the loss curve
plt.plot(mlp.loss_curve_, color='red')
plt.title('Loss Curve during Training')
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.show()
```
